### Renaming some of the columns

```python
def rename_columns(df):
    df.rename(columns={'sex': 'gender'}, inplace=True)
    df.rename(columns={'q15hiedu': 'education'}, inplace=True)
    df.rename(columns={'population': 'race'}, inplace=True)
    df.rename(columns={'age_grp1': 'age_grp'}, inplace=True)
    df.rename(columns={'employ_status1': 'employ_status'}, inplace=True)
    return df
```

### Mapping Some Numerical Columns To Categorical

Some of the features which were catergorical were label encoded.  We want to return the values to catergorical for better analysis 

```python
# Define mapping dictionaries based on the provided key
education_map = {
    0: 'Grade R/0',
    1: 'Grade 1/Sub A/Class 1',
    2: 'Grade 2/Sub B/Class 2',
    3: 'Grade 3/Standard 1/ABET/AET 1',
    4: 'Grade 4/Standard 2',
    5: 'Grade 5/Standard 3/ABET/AET 2',
    6: 'Grade 6/Standard 4',
    7: 'Grade 7/Standard 5/ABET/AET 3',
    8: 'Grade 8/Standard 6/Form 1',
    9: 'Grade 9/Standard 7/Form 2/ABET/AET 4/NCV Level 1/Occupational Certificate-NQF Level 1',
    10: 'Grade 10/Standard 8/Form 3/NCV Level 2/Occupational Certificate-NQF Level 2',
    11: 'Grade 11/Standard 9/Form 4/NCV Level 3/Occupational Certificate-NQF Level 3',
    12: 'Grade 12/Standard 10/Form 5/National Senior Certificate/Matric/ NCV Level 4/Occupational Certificate-NQF Level 4',
    13: 'NTC l/N1',
    14: 'NTC ll/N2',
    15: 'NTC lll/N3',
    16: 'N4/NTC 4/Occupational Certificate-NQF Level 5',
    17: 'N5/NTC 5/Occupational Certificate-NQF Level 5',
    18: 'N6/NTC 6/Occupational Certificate-NQF Level 5',
    19: 'Certificate with less than Grade 12/Standard 10',
    20: 'Diploma with less than Grade 12/Standard 10',
    21: 'Higher/National/Advance certificate with Grade 12/Std 10/Occupational Certificate-NQF Level 5',
    22: 'Diploma with Grade 12/Standard 10/Occupational Certificate-NQF Level 6',
    23: 'Higher Diploma/Occupational Certificate (B-Tech Diploma)-NQF Level 7',
    24: 'Post Higher Diploma (Masters Diploma and Masters Degree)-NQF Level 9',
    25: 'Bachelors Degree/Occupational Certificate-NQF Level 7',
    26: 'Honours Degree/Postgraduate Diploma/Occupational Certificate-NQF Level 8',
    27: 'Doctoral Degrees (Doctoral Diploma and PhD)-NQF Level 10',
    28: 'Other',
    29: 'Do not know',
    98: 'No schooling',
    99: 'Unspecified'
}

age_grp_map = {
    1: '00-04', 2: '05-09', 3: '10-14', 4: '15-19', 5: '20-24', 6: '25-29', 7: '30-34', 8: '35-39',
    9: '40-44', 10: '35-39', 11: '40-44', 12: '45-49', 13: '50-54', 14: '55-59', 15: '60-64', 16:'75+'
}

prov_map = {
    1: 'Western Cape', 2: 'Eastern Cape', 3: 'Northern Cape', 4: 'Free State', 5: 'KwaZulu-Natal',
    6: 'North West', 7: 'Gauteng', 8: 'Mpumalanga', 9: 'Limpopo'
}

race_map = {
    1: 'African/Black', 2: 'Coloured', 3: 'Indian/Asian', 4: 'Caucasian/White'
}

gender_map = {
    1: 'Male', 2: 'Female'
}

employ_status_map = {
    1: 'Employed', 2: 'Unemployed', 8: 'Not Applicable', 9: 'Unspecified'
}

disab_map = {
    0: 'Not Disabled', 1: 'Disabled', 9: 'Unspecified'
}

# Map numerical values to categorical values using the dictionaries
def map_numerical_to_categorical(df):
    df['education'] = df['education'].map(education_map)
    df['age_grp'] = df['age_grp'].map(age_grp_map)
    df['prov'] = df['prov'].map(prov_map)
    df['race'] = df['race'].map(race_map)
    df['gender'] = df['gender'].map(gender_map)
    df['employ_status'] = df['employ_status'].map(employ_status_map)
    df['disab'] = df['disab'].map(disab_map)
    return df
```

```python
dataset_1 = map_numerical_to_categorical(dataset_1)
dataset_2 = map_numerical_to_categorical(dataset_2)
dataset_3 = map_numerical_to_categorical(dataset_3)
dataset_4 = map_numerical_to_categorical(dataset_4)
dataset_5 = map_numerical_to_categorical(dataset_5)
dataset_6 = map_numerical_to_categorical(dataset_6)
dataset_7 = map_numerical_to_categorical(dataset_7)
```

### Filter The Dataset by age

Here we want to filter the dataset by the youth age in South Africa, which is from 18 to 50

```python
def filter_datasets_by_age(df):
    # Filter the dataset by age between 18 and 50
    filtered_df = df[(df['age'] >= 18) & (df['age'] <= 50)]
    return filtered_df
```

```python
dataset_1 = filter_datasets_by_age(dataset_1)
dataset_2 = filter_datasets_by_age(dataset_2)
dataset_3 = filter_datasets_by_age(dataset_3)
dataset_4 = filter_datasets_by_age(dataset_4)
dataset_5 = filter_datasets_by_age(dataset_5)
dataset_6 = filter_datasets_by_age(dataset_6)
dataset_7 = filter_datasets_by_age(dataset_7)
```

```python
list_of_datasets=[dataset_1, dataset_2, dataset_3, dataset_4, dataset_5, dataset_6, dataset_7]
```

### Merging All The Datasets

```python
# Merge all datasets in the datasets list
merged_data = pd.concat(list_of_datasets)

# Optionally, you can reset the index of the merged data
merged_data.reset_index(drop=True, inplace=True)
```

### Removing Duplicate

Here we are going to remove duplicates based on **Uqnr** and **Personnr**

```python
# Remove duplicates based on 'uqnr' and 'personnr' columns
master_dataset = merged_data.drop_duplicates(subset=['uqnr', 'personnr'])
```
